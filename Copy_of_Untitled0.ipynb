```python
import urllib.request
hand = urllib.request.urlopen('https://en.wikipedia.org/wiki/Henry_Ford')

import re
def ascii_histogram(seq) -> None: #create a histogram from a list of numbers
  counted = count_elements(seq) #generate frequency table (dictionary)
  for key in sorted(counted):
    plus_str = '+' * counted[key]
    print(f'{key} {plus_str}') #prints out asterisk
def count_elements(seq): #creates a dictionary
    hist = dict()
    for date in seq:
      hist[date]=hist.get(date,0)+ 1
    return hist

dates = []
for line in hand:
    line = line.decode().strip()
    found_years = re.findall(r'\b(1[0-9]{3}|2[0-9]{3})\b', line)
    # Add found years to the dates list
    dates.extend(found_years)

# Generate and display the ASCII histogram
print("Year Frequency Histogram:")
ascii_histogram(dates)













```
